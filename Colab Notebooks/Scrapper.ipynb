```python
import numpy as np
def read_data(dataset_name):
    p = list(np.load(f"/content/drive/MyDrive/parrot/chessevaldata/{dataset_name}_pos.npy"))
    f = list(np.load(f"/content/drive/MyDrive/parrot/chessevaldata/{dataset_name}_feat.npy"))
    e = list(np.load(f"/content/drive/MyDrive/parrot/chessevaldata/{dataset_name}_eval.npy"))
    return (p, f, e)

def read_pos(dataset_name):
    p = list(np.load(f"/content/drive/MyDrive/parrot/chessevaldata/{dataset_name}_pos.npy"))
    return p
```

```python
import re

chr_to_num = {"r": 0, "R": 1, "n": 2, "N": 3, "b": 4, "B": 5, "k": 6, "K": 7, "q": 8, "Q": 9, "p": 10, "P": 11}

def square_to_int(sq):
    return (ord(sq[0]) - 97) * 8 + int(sq[1]) - 1

def squareint_to_square(sqint):
    return (sqint // 8, sqint % 8)

def int_to_bin(anint, pad=4):
    return [int(_) for _ in "0" * (pad - len(bin(anint)[2:])) + bin(anint)[2:]]

def board_to_boardlist(board):
    boards = [[[0 for _ in range(8)] for _ in range(8)] for _ in range(12)]
    for square, piece in board.piece_map().items():
        idx = squareint_to_square(square)
        boards[chr_to_num[piece.symbol()]][idx[0]][idx[1]] = 1
    return boards

def board_to_boardmap(board):
    boards = [[0 for _ in range(8)] for _ in range(8)]
    for square, piece in board.piece_map().items():
        idx = squareint_to_square(square)
        boards[idx[0]][idx[1]] = round((chr_to_num[piece.symbol()] + 1) / 12, 4)
    return [boards]


def fen_to_feature_wboard_list(fen_str):
    board = re.split(" |/", fen_str)
    r = 0
    whosemove = []
    enpassqnum = board[10]
    can_enpassant = [0]
    castling_rights = [0, 0, 0, 0]
    if enpassqnum != "-":
        enpassqnum = int_to_bin(square_to_int(enpassqnum), pad=6)
        can_enpassant = [1]
    else:
        enpassqnum = int_to_bin(0, pad=6)
    for entry in board:
        if entry == "w":
            whosemove = [1]
        elif entry == "b":
            whosemove = [0]
        elif r == 9:
            if entry.find("K") != -1:
                castling_rights[0] = 1
            if entry.find("Q") != -1:
                castling_rights[1] = 1
            if entry.find("k") != -1:
                castling_rights[2] = 1
            if entry.find("q") != -1:
                castling_rights[3] = 1
        r += 1
    return whosemove + can_enpassant + enpassqnum + castling_rights
```

```python
import numpy as np

your_filename = "/content/drive/MyDrive/parrot/lichess_db_eval.jsonl.zst"
dctx = zstd.ZstdDecompressor()
l = 0
num_datasets = 0
num_sub_datasets = 0
pos_list = []
feat_list = []
eval_list = []
with open(your_filename, 'rb') as compressed:
    with dctx.stream_reader(compressed) as reader:
        text_stream = io.TextIOWrapper(reader, encoding='utf-8')
        for line in text_stream:
            data = json.loads(line)
            max_depth = -1
            turn = (data["fen"].find("w") != -1)
            best_eval = None
            for evals in data["evals"]:
                if evals["depth"] > max_depth:
                    max_depth = evals["depth"]
                    best_score = -1000000 if turn else 10000000
                    for variation in evals["pvs"]:
                        try:
                            score = 0.4 * ((1 - np.exp(-variation["cp"] / 200)) / (1 + np.exp(-variation["cp"] / 200))) + 0.5
                        except:
                            score = 0.9 + (10 - variation["mate"]) / 100 if variation["mate"] > 0 else 0.1 + (-variation["mate"] - 10) / 100
                            if score < 0.9 and variation["mate"] > 0:
                                score = 0.9
                            elif score > 0.1 and variation["mate"] < 0:
                                score = 0.1
                        if turn:
                            if score > best_score:
                                best_score = score
                        elif not turn:
                            if score < best_score:
                                best_score = score
                    best_eval = best_score
            pos_list.append(board_to_boardmap(chess.Board(data["fen"])))
            feat_list.append(fen_to_feature_wboard_list(data["fen"]))
            eval_list.append(best_eval)
            if l % 10000 == 0:
                print(l)
            if l % 65536 == 65535:
                np.save(f"/content/drive/MyDrive/parrot/chessevaldata/train_{num_datasets}_{num_sub_datasets}_pos.npy", np.array(pos_list))
                np.save(f"/content/drive/MyDrive/parrot/chessevaldata/train_{num_datasets}_{num_sub_datasets}_feat.npy", np.array(feat_list))
                np.save(f"/content/drive/MyDrive/parrot/chessevaldata/train_{num_datasets}_{num_sub_datasets}_eval.npy", np.array(eval_list))
                print("Saved training dataset ", num_datasets, num_sub_datasets)
                pos_list = []
                feat_list = []
                eval_list = []
                num_sub_datasets += 1
                if num_sub_datasets == 16:
                    num_sub_datasets = 0
                    num_datasets += 1
            l += 1
```
